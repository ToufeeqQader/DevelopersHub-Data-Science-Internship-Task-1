```python
"""
Exploring and Visualizing a Simple Dataset.
The objective here is to learn how to read, summarize, and visualize data.
We will be using the famous Iris Dataset, which we can easily load using the seaborn library.
"""
```

```python
# Import the necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset directly from seaborn
df = sns.load_dataset('iris')

print("Dataset Shape:", df.shape)
print("Columns:", df.columns)
```

```python
display(df.head())
```

```python
# Scatter Plot
```

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.title('Relationship: Sepal Length vs Sepal Width')
plt.show()
```

```python
# Histogram
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='petal_length', kde=True, hue='species')
plt.title('Data Distribution of Petal Length')
plt.show()
```

```python
# Box Plot
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='species', y='sepal_width')
plt.title('Spread and Outliers: Box Plot of Sepal Width by Species')
plt.show()
```
